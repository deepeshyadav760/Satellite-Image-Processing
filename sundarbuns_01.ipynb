```python
import ee
import geemap
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
import ipywidgets as widgets
from IPython.display import display, HTML

# Initialize Earth Engine with Project ID
Project_id = 'ee-deepeshy'
try:
    ee.Initialize(project=Project_id)
    print("Earth Engine Initialised Successfully!")
except ee.EEException:
    print("Authentication required. Please follow the instructions.")
    ee.Authenticate()
    ee.Initialize(project=Project_id)



class SundarbansAnalysis:
    def __init__(self):
        # Define constants for carbon estimation
        self.carbon_constants = {
            'agb_to_bgb_ratio': 0.39,  # Root-to-shoot ratio for mangroves
            'carbon_to_biomass_ratio': 0.47,  # Carbon content in biomass
            'soil_carbon_depth_cm': 100,  # Depth for soil carbon calculation in cm
            'soil_carbon_density': 0.03,  # g/cm³ for top 1m in mangroves
            'water_carbon_factor': 0.005,  # tC/ha for water bodies with mangrove influence
            'sequestration_rate': 10,  # tonnes CO2/ha/yr
            'mangrove_carbon_density': 447,  # tC/ha (literature value for Sundarbans)
        }

        # Define Sundarbans region
        self.define_region()

        # Initialize results dictionary
        self.results = {}

    def define_region(self):
        """Define the Sundarbans region of interest"""
        # Get West Bengal boundary
        self.west_bengal = ee.FeatureCollection("FAO/GAUL/2015/level1").filter(ee.Filter.eq('ADM1_NAME', 'West Bengal'))

        # Create a more precise bounding box for Sundarbans
        self.sundarbans_geo = ee.Geometry.Rectangle([88.0, 21.5, 89.5, 22.20])

        # Get the intersection
        self.sundarbans = self.west_bengal.geometry().intersection(self.sundarbans_geo)

        # Scale for area calculations - 30m to match Landsat resolution
        self.scale = 30

    def get_landsat_collection(self, start_date, end_date, cloud_cover_max=30):
        """Get Landsat 8/9 collection for the specified date range"""
        # Try Landsat 9 first (launched in 2021)
        landsat9_collection = ee.ImageCollection("LANDSAT/LC09/C02/T1_L2") \
            .filterDate(start_date, end_date) \
            .filterBounds(self.sundarbans) \
            .filter(ee.Filter.lt('CLOUD_COVER', cloud_cover_max))

        # Landsat 8 collection
        landsat8_collection = ee.ImageCollection("LANDSAT/LC08/C02/T1_L2") \
            .filterDate(start_date, end_date) \
            .filterBounds(self.sundarbans) \
            .filter(ee.Filter.lt('CLOUD_COVER', cloud_cover_max))

        # Merge collections (will use L9 when available, L8 otherwise)
        collection = ee.ImageCollection(landsat9_collection.merge(landsat8_collection))

        # Check if collection is empty and provide a helpful message
        collection_size = collection.size().getInfo()
        if collection_size == 0:
            print(f"Warning: No Landsat images found for the period {start_date} to {end_date} with cloud cover < {cloud_cover_max}%")
            print("Try expanding the date range or increasing the maximum cloud cover threshold")
            return None

        # Sort by cloud cover and return
        return collection.sort('CLOUD_COVER')

    def apply_cloud_mask(self, image):
        """Apply cloud and shadow mask using QA_PIXEL band"""
        # Get the QA band
        qa = image.select('QA_PIXEL')

        # Bits 3 and 4 are cloud shadow and cloud, respectively
        cloud_shadow_bit_mask = 1 << 3
        cloud_bit_mask = 1 << 4

        # Create mask
        mask = qa.bitwiseAnd(cloud_shadow_bit_mask).eq(0) \
            .And(qa.bitwiseAnd(cloud_bit_mask).eq(0))

        # Select the bands we need and apply scaling
        blue = image.select('SR_B2').multiply(0.0000275).add(-0.2).rename('Blue')
        green = image.select('SR_B3').multiply(0.0000275).add(-0.2).rename('Green')
        red = image.select('SR_B4').multiply(0.0000275).add(-0.2).rename('Red')
        nir = image.select('SR_B5').multiply(0.0000275).add(-0.2).rename('NIR')
        swir1 = image.select('SR_B6').multiply(0.0000275).add(-0.2).rename('SWIR1')
        swir2 = image.select('SR_B7').multiply(0.0000275).add(-0.2).rename('SWIR2')

        # Create a new image with the scaled bands
        scaled_image = ee.Image.cat([blue, green, red, nir, swir1, swir2])

        # Apply mask
        return scaled_image.updateMask(mask)

    def calculate_indices(self, image):
        """Calculate all vegetation and water indices"""
        # NDVI (Normalized Difference Vegetation Index)
        ndvi = image.normalizedDifference(['NIR', 'Red']).rename('NDVI')

        # EVI (Enhanced Vegetation Index)
        # EVI = 2.5 * ((NIR - Red) / (NIR + 6 * Red - 7.5 * Blue + 1))
        evi = image.expression(
            '2.5 * ((NIR - RED) / (NIR + 6 * RED - 7.5 * BLUE + 1))', {
                'NIR': image.select('NIR'),
                'RED': image.select('Red'),
                'BLUE': image.select('Blue')
            }
        ).rename('EVI')

        # EVI2 (Enhanced Vegetation Index 2)
        # EVI2 = 2.5 * ((NIR - Red) / (NIR + 2.4 * Red + 1))
        evi2 = image.expression(
            '2.5 * ((NIR - RED) / (NIR + 2.4 * RED + 1))', {
                'NIR': image.select('NIR'),
                'RED': image.select('Red')
            }
        ).rename('EVI2')

        # NDMI (Normalized Difference Moisture Index)
        # NDMI = (NIR - SWIR1) / (NIR + SWIR1)
        ndmi = image.normalizedDifference(['NIR', 'SWIR1']).rename('NDMI')

        # NDWI (Normalized Difference Water Index)
        # NDWI = (Green - NIR) / (Green + NIR)
        ndwi = image.normalizedDifference(['Green', 'NIR']).rename('NDWI')

        # MNDWI (Modified Normalized Difference Water Index)
        # MNDWI = (Green - SWIR1) / (Green + SWIR1)
        mndwi = image.normalizedDifference(['Green', 'SWIR1']).rename('MNDWI')

        # Add all indices to the original image
        return image.addBands([ndvi, evi, evi2, ndmi, ndwi, mndwi])

    def create_annual_composite(self, year):
        """Create a seasonal composite for a given year"""
        # Define start and end dates for the year
        start_date = f"{year}-01-01"
        end_date = f"{year}-12-31"

        # Get Landsat collection
        collection = self.get_landsat_collection(start_date, end_date)

        if collection is None:
            return None

        # Apply cloud mask to all images
        masked_collection = collection.map(self.apply_cloud_mask)

        # Create a median composite
        composite = masked_collection.median().clip(self.sundarbans)

        # Calculate indices
        indexed_composite = self.calculate_indices(composite)

        return indexed_composite

    def create_seasonal_composites(self, year):
        """Create seasonal composites for a given year"""
        seasons = {
            'winter': [f"{year-1}-12-01", f"{year}-02-28"],
            'summer': [f"{year}-03-01", f"{year}-05-31"],
            'monsoon': [f"{year}-06-01", f"{year}-09-30"],
            'post_monsoon': [f"{year}-10-01", f"{year}-11-30"]
        }

        seasonal_composites = {}

        for season, dates in seasons.items():
            # Get Landsat collection for the season
            collection = self.get_landsat_collection(dates[0], dates[1])

            if collection is None or collection.size().getInfo() == 0:
                print(f"No images available for {season} {year}")
                seasonal_composites[season] = None
                continue

            # Apply cloud mask to all images
            masked_collection = collection.map(self.apply_cloud_mask)

            # Create a median composite
            composite = masked_collection.median().clip(self.sundarbans)

            # Calculate indices
            indexed_composite = self.calculate_indices(composite)

            seasonal_composites[season] = indexed_composite

        return seasonal_composites

    def identify_mangroves(self, composite):
        """Identify potential mangrove areas using indices and ancillary data"""
        if composite is None:
            return None

        # Extract indices
        ndvi = composite.select('NDVI')
        ndwi = composite.select('NDWI')
        mndwi = composite.select('MNDWI')

        # Use ESA WorldCover for land cover at 10m resolution
        landcover = ee.ImageCollection("ESA/WorldCover/v200").first().clip(self.sundarbans)

        # Extract tree cover (class 10) and wetland (class 90) as potential mangrove areas
        potential_mangrove_areas = landcover.eq(10).Or(landcover.eq(90))

        # Use elevation data for coastal mask
        elevation = ee.Image("USGS/SRTMGL1_003").clip(self.sundarbans)
        coastal_mask = elevation.lte(20)  # Areas less than 20m above sea level

        # Combine indices and masks to identify mangroves
        # High NDVI (>0.5), negative NDWI (not open water), and in potential mangrove areas near coast
        mangrove_mask = ndvi.gt(0.5) \
            .And(ndwi.lt(0.0)) \
            .And(potential_mangrove_areas) \
            .And(coastal_mask)

        return mangrove_mask

    def estimate_carbon_pools(self, composite, year):
        """Estimate carbon pools for a given composite"""
        if composite is None:
            return None

        # Identify mangroves
        mangrove_mask = self.identify_mangroves(composite)

        # Get vegetation indices
        ndvi = composite.select('NDVI')
        evi = composite.select('EVI')

        # 1. Above-ground biomass (AGB) estimation
        agb = ndvi.expression(
            '140 * exp(1.7 * NDVI)', {'NDVI': ndvi}
        ).multiply(mangrove_mask).rename('AGB')

        # 2. Below-ground biomass (BGB) estimation
        bgb = agb.multiply(self.carbon_constants['agb_to_bgb_ratio']).rename('BGB')

        # 3. Total biomass carbon
        biomass_carbon = agb.add(bgb).multiply(self.carbon_constants['carbon_to_biomass_ratio']).rename('Biomass_Carbon')

        # 4. Soil carbon estimation
        # Use ISRIC SoilGrids for soil organic carbon content (summing layers up to 100cm)
        soil_carbon_data = ee.Image("projects/soilgrids-isric/soc_mean") \
            .select(['soc_0-5cm_mean', 'soc_5-15cm_mean', 'soc_15-30cm_mean',
                     'soc_30-60cm_mean', 'soc_60-100cm_mean']) \
            .reduce(ee.Reducer.sum()) \
            .clip(self.sundarbans)

        # Convert from cg/kg to tC/ha (adjust the conversion factors as needed)
        # Assuming soil_carbon_data is in cg/kg, conversion steps:
        # 1 cg/kg = 0.01 g/kg
        # Assuming bulk density of 1.5 g/cm³ and depth 100cm (1m)
        # Conversion factor: (0.01 g/kg) * (1.5 g/cm³) * (100 cm) * (1 kg / 1000 g) * (10^4 m²/ha)
        # = 15 tC/ha
        soil_carbon = soil_carbon_data.multiply(0.01) \
            .multiply(1.5) \
            .multiply(100) \
            .divide(1000) \
            .multiply(10000) \
            .multiply(mangrove_mask) \
            .rename('Soil_Carbon')

        # 5. Water carbon estimation using MNDWI
        mndwi = composite.select('MNDWI')
        water_mask = mndwi.gt(0.2)  # Water bodies with mangrove influence
        water_carbon = water_mask.multiply(self.carbon_constants['water_carbon_factor']).rename('Water_Carbon')

        # 6. Total carbon
        total_carbon = biomass_carbon.add(soil_carbon).add(water_carbon).rename('Total_Carbon')

        # Combine all carbon pools into a single image
        carbon_pools = ee.Image.cat([
            biomass_carbon, soil_carbon, water_carbon, total_carbon
        ]).set('year', year)

        return carbon_pools, mangrove_mask

    def calculate_area_statistics(self, mangrove_mask, year):
        """Calculate area statistics for mangroves"""
        if mangrove_mask is None:
            return None

        # Calculate total area of the intersection
        total_area_image = ee.Image.pixelArea().clip(self.sundarbans).divide(10000)  # m² → ha
        total_area_ha = total_area_image.reduceRegion(
            reducer=ee.Reducer.sum(),
            geometry=self.sundarbans,
            scale=self.scale,
            maxPixels=1e12
        ).get('area').getInfo()

        # Calculate mangrove area in hectares
        mangrove_area_image = mangrove_mask.multiply(ee.Image.pixelArea()).divide(10000)  # m² → ha
        mangrove_area_ha = mangrove_area_image.reduceRegion(
            reducer=ee.Reducer.sum(),
            geometry=self.sundarbans,
            scale=self.scale,
            maxPixels=1e12
        ).get(mangrove_mask.bandNames().get(0)).getInfo()

        # Calculate carbon stock and sequestration
        total_carbon_stock = mangrove_area_ha * self.carbon_constants['mangrove_carbon_density']
        total_co2_seq = mangrove_area_ha * self.carbon_constants['sequestration_rate']

        return {
            'year': year,
            'total_area_ha': total_area_ha,
            'mangrove_area_ha': mangrove_area_ha,
            'mangrove_percentage': (mangrove_area_ha / total_area_ha) * 100,
            'total_carbon_stock': total_carbon_stock,
            'total_co2_seq': total_co2_seq
        }

    def calculate_carbon_statistics(self, carbon_pools, year):
        """Calculate carbon statistics from carbon pool image"""
        if carbon_pools is None:
            return None

        # Calculate statistics for each carbon pool
        carbon_stats = carbon_pools.reduceRegion(
            reducer=ee.Reducer.sum(),
            geometry=self.sundarbans,
            scale=self.scale,
            maxPixels=1e12
        ).getInfo()

        # Add year information
        carbon_stats['year'] = year

        return carbon_stats

    def analyze_historical_trends(self, start_year=2018, end_year=2024):
        """Analyze historical carbon trends from start_year to end_year"""
        print(f"Analyzing historical trends from {start_year} to {end_year}...")

        # Storage for results
        yearly_results = []
        carbon_pool_images = []

        # Process each year
        for year in range(start_year, end_year + 1):
            print(f"Processing year {year}...")

            # Create annual composite
            composite = self.create_annual_composite(year)

            if composite is None:
                print(f"No data available for {year}, skipping.")
                continue

            # Estimate carbon pools
            carbon_pools, mangrove_mask = self.estimate_carbon_pools(composite, year)

            # Calculate statistics
            area_stats = self.calculate_area_statistics(mangrove_mask, year)
            carbon_stats = self.calculate_carbon_statistics(carbon_pools, year)

            # Combine statistics
            year_result = {**area_stats, **carbon_stats}
            yearly_results.append(year_result)

            # Store carbon pool image with year as property
            carbon_pool_images.append(carbon_pools)

        # Convert to pandas DataFrame for easier analysis
        self.results['historical_trends'] = pd.DataFrame(yearly_results)

        # Create an ImageCollection for carbon pools
        self.results['carbon_pool_collection'] = ee.ImageCollection.fromImages(carbon_pool_images)

        return self.results['historical_trends']

    def predict_future_carbon(self, prediction_years=5):
        """Predict future carbon trends using historical data"""
        print(f"Predicting carbon trends for the next {prediction_years} years...")

        # Check if historical analysis has been performed
        if 'historical_trends' not in self.results:
            print("Error: Historical analysis must be performed before prediction.")
            return None

        # Get historical data
        historical_data = self.results['historical_trends']

        if len(historical_data) < 2:
            print("Error: Insufficient historical data for prediction (need at least 2 years).")
            return None

        # Prepare data for prediction
        X = historical_data['year'].values.reshape(-1, 1)

        # Predictions to make
        targets = ['mangrove_area_ha', 'total_carbon_stock', 'total_co2_seq',
                  'Biomass_Carbon', 'Soil_Carbon', 'Water_Carbon', 'Total_Carbon']

        # Storage for predictions
        predictions = []

        # Last year in historical data
        last_year = historical_data['year'].max()

        # Create prediction years
        future_years = range(last_year + 1, last_year + prediction_years + 1)
        X_future = np.array(list(future_years)).reshape(-1, 1)

        # Fit regression model for each target
        for target in targets:
            if target in historical_data.columns:
                y = historical_data[target].values

                # Use RandomForest for prediction if enough data, otherwise linear regression
                if len(y) >= 5:
                    model = RandomForestRegressor(n_estimators=100, random_state=42)
                else:
                    model = LinearRegression()

                model.fit(X, y)

                # Make predictions
                y_future = model.predict(X_future)

                # Add to predictions DataFrame
                for i, year in enumerate(future_years):
                    while len(predictions) <= i:
                        predictions.append({'year': year})

                    predictions[i][target] = y_future[i]

        # Convert to DataFrame
        future_df = pd.DataFrame(predictions)

        # Store results
        self.results['future_predictions'] = future_df

        # Combine historical and future for a complete timeline
        self.results['complete_timeline'] = pd.concat([historical_data, future_df], ignore_index=True)

        return future_df

    def detect_changes(self, base_year, compare_year):
        """Detect changes in carbon stocks between two years"""
        print(f"Detecting changes between {base_year} and {compare_year}...")

        # Check if we have the necessary years in our carbon pool collection
        if 'carbon_pool_collection' not in self.results:
            print("Error: Historical analysis must be performed first.")
            return None

        # Get carbon pools for both years
        carbon_collection = self.results['carbon_pool_collection']

        base_carbon = carbon_collection.filter(ee.Filter.eq('year', base_year)).first()
        compare_carbon = carbon_collection.filter(ee.Filter.eq('year', compare_year)).first()

        # Calculate difference (compare - base)
        carbon_difference = compare_carbon.subtract(base_carbon).rename([
            'Biomass_Carbon_Diff', 'Soil_Carbon_Diff', 'Water_Carbon_Diff', 'Total_Carbon_Diff'
        ])

        # Store result
        self.results['change_detection'] = {
            'base_year': base_year,
            'compare_year': compare_year,
            'difference_image': carbon_difference
        }

        return carbon_difference

    def visualize_carbon_map(self, year=None, carbon_type='Total_Carbon'):
        """Create a visualization map for carbon distribution"""
        if 'carbon_pool_collection' not in self.results:
            print("Error: Historical analysis must be performed first.")
            return None

        # Get the carbon pool image for the specified year, or the most recent one
        carbon_collection = self.results['carbon_pool_collection']

        if year is not None:
            carbon_image = carbon_collection.filter(ee.Filter.eq('year', year)).first()
            title_year = str(year)
        else:
            # Get the most recent year
            year_list = carbon_collection.aggregate_array('year').sort().getInfo()
            year = year_list[-1]
            carbon_image = carbon_collection.filter(ee.Filter.eq('year', year)).first()
            title_year = str(year) + " (most recent)"

        # Create visualization map
        map1 = geemap.Map()
        map1.centerObject(self.sundarbans, 10)

        # Add base layers
        map1.addLayer(self.west_bengal.geometry(), {'color': 'blue'}, "West Bengal", False)
        map1.addLayer(self.sundarbans, {'color': 'red'}, "Sundarbans Region")

        # Add carbon layer with appropriate visualization
        if carbon_type == 'Total_Carbon':
            viz_params = {'min': 0, 'max': 500, 'palette': ['yellow', 'orange', 'darkgreen']}
            title = f"Total Carbon Stock ({title_year})"
        elif carbon_type == 'Biomass_Carbon':
            viz_params = {'min': 0, 'max': 200, 'palette': ['yellow', 'green']}
            title = f"Biomass Carbon ({title_year})"
        elif carbon_type == 'Soil_Carbon':
            viz_params = {'min': 0, 'max': 300, 'palette': ['brown', 'darkbrown']}
            title = f"Soil Carbon ({title_year})"
        elif carbon_type == 'Water_Carbon':
            viz_params = {'min': 0, 'max': 10, 'palette': ['cyan', 'blue']}
            title = f"Water Carbon ({title_year})"

        map1.addLayer(carbon_image.select(carbon_type), viz_params, title)

        # Add a legend
        map1.add_colorbar(viz_params, label=f"{carbon_type} (tC/ha)")

        # Get statistics for the info panel
        if 'historical_trends' in self.results:
            year_data = self.results['historical_trends'][self.results['historical_trends']['year'] == year]

            if not year_data.empty:
                stats_html = f"""
                <div style="background-color: rgba(255, 255, 255, 0.8); padding: 10px; border-radius: 5px;">
                    <h3>Carbon Analysis for {year}</h3>
                    <p><b>Mangrove Area:</b> {year_data['mangrove_area_ha'].values[0]:,.0f} hectares
                       ({year_data['mangrove_percentage'].values[0]:.1f}%)</p>
                    <p><b>Total Carbon Stock:</b> {year_data['total_carbon_stock'].values[0]:,.0f} tonnes C</p>
                    <p><b>Annual CO2 Absorption:</b> {year_data['total_co2_seq'].values[0]:,.0f} tonnes CO2/year</p>
                </div>
                """
                map1.add_html(stats_html, position="bottomleft")

        return map1

    def visualize_change_map(self):
        """Visualize the change in carbon stocks between two years"""
        if 'change_detection' not in self.results:
            print("Error: Change detection must be performed first.")
            return None

        # Get change detection results
        change_detection = self.results['change_detection']
        difference_image = change_detection['difference_image']
        base_year = change_detection['base_year']
        compare_year = change_detection['compare_year']

        # Create visualization map
        map1 = geemap.Map()
        map1.centerObject(self.sundarbans, 10)

        # Add base layers
        map1.addLayer(self.sundarbans, {'color': 'gray'}, "Sundarbans Region", False)

        # Add change layers
        # For Total Carbon change
        viz_params = {
            'min': -100,
            'max': 100,
            'palette': ['red', 'white', 'green']
        }

        map1.addLayer(
            difference_image.select('Total_Carbon_Diff'),
            viz_params,
            f"Total Carbon Change ({base_year} to {compare_year})"
        )

        # Add a legend
        map1.add_colorbar(viz_params, label=f"Carbon Change (tC/ha)")

        # Add info panel
        info_html = f"""
        <div style="background-color: rgba(255, 255, 255, 0.8); padding: 10px; border-radius: 5px;">
            <h3>Carbon Change Analysis</h3>
            <p><b>Base Year:</b> {base_year}</p>
            <p><b>Compare Year:</b> {compare_year}</p>
            <p><b>Legend:</b></p>
            <ul>
                <li><span style="color: red;">■</span> Carbon Loss</li>
                <li><span style="color: white;">■</span> No Change</li>
                <li><span style="color: green;">■</span> Carbon Gain</li>
            </ul>
        </div>
        """
        map1.add_html(info_html, position="bottomright")

        return map1

    def create_time_series_charts(self):
        """Create time series charts for carbon trends"""
        if 'complete_timeline' not in self.results:
            print("Error: Historical analysis and prediction must be performed first.")
            return None

        # Get complete timeline data
        data = self.results['complete_timeline']

        # Determine historical vs prediction cutoff
        historical_end = self.results['historical_trends']['year'].max()

        # Setup plotting
        plt.figure(figsize=(15, 10))

        # Plot 1: Mangrove Area
        plt.subplot(2, 2, 1)
        plt.plot(data[data['year'] <= historical_end]['year'],
                 data[data['year'] <= historical_end]['mangrove_area_ha'],
                 'o-', color='green', label='Historical')

        if any(data['year'] > historical_end):
            plt.plot(data[data['year'] > historical_end]['year'],
                     data[data['year'] > historical_end]['mangrove_area_ha'],
                     'o--', color='lightgreen', label='Predicted')

        plt.title('Mangrove Area (ha) Over Time')
        plt.xlabel('Year')
        plt.ylabel('Area (hectares)')
        plt.grid(True, alpha=0.3)
        plt.legend()

        # Plot 2: Total Carbon Stock
        plt.subplot(2, 2, 2)
        plt.plot(data[data['year'] <= historical_end]['year'],
                 data[data['year'] <= historical_end]['total_carbon_stock'],
                 'o-', color='blue', label='Historical')

        if any(data['year'] > historical_end):
            plt.plot(data[data['year'] > historical_end]['year'],
                     data[data['year'] > historical_end]['total_carbon_stock'],
                     'o--', color='lightblue', label='Predicted')

        plt.title('Total Carbon Stock (tonnes C) Over Time')
        plt.xlabel('Year')
        plt.ylabel('Carbon Stock (tonnes C)')
        plt.grid(True, alpha=0.3)
        plt.legend()

        # Plot 3: Annual CO2 Absorption
        plt.subplot(2, 2, 3)
        plt.plot(data[data['year'] <= historical_end]['year'],
                 data[data['year'] <= historical_end]['total_co2_seq'],
                 'o-', color='purple', label='Historical')

        if any(data['year'] > historical_end):
            plt.plot(data[data['year'] > historical_end]['year'],
                     data[data['year'] > historical_end]['total_co2_seq'],
                     'o--', color='plum', label='Predicted')

        plt.title('Annual CO2 Absorption (tonnes CO2/year) Over Time')
        plt.xlabel('Year')
        plt.ylabel('CO2 Absorption (tonnes/year)')
        plt.grid(True, alpha=0.3)
        plt.legend()

        # Plot 4: Carbon Pools
        plt.subplot(2, 2, 4)

        # Check which carbon pool columns are available
        carbon_cols = [col for col in ['Biomass_Carbon', 'Soil_Carbon', 'Water_Carbon']
                       if col in data.columns]

        for col, color in zip(carbon_cols, ['green', 'brown', 'blue']):
            hist_data = data[data['year'] <= historical_end]
            plt.plot(hist_data['year'], hist_data[col], 'o-', color=color, label=f'Historical {col}')

            if any(data['year'] > historical_end):
                pred_data = data[data['year'] > historical_end]
                plt.plot(pred_data['year'], pred_data[col], 'o--', color=color, alpha=0.5, label=f'Predicted {col}')

        plt.title('Carbon Pools Over Time')
        plt.xlabel('Year')
        plt.ylabel('Carbon (tonnes)')
        plt.grid(True, alpha=0.3)
        plt.legend()

        plt.tight_layout()

        return plt

    def create_comparative_charts(self, year=None):
        """Create comparative charts for carbon pool distribution"""
        if 'historical_trends' not in self.results:
            print("Error: Historical analysis must be performed first.")
            return None

        # Get data for the most recent year if not specified
        data = self.results['historical_trends']

        if year is None:
            year = data['year'].max()

        year_data = data[data['year'] == year]

        if year_data.empty:
            print(f"Error: No data available for year {year}.")
            return None

        # Check which carbon pool columns are available
        carbon_cols = [col for col in ['Biomass_Carbon', 'Soil_Carbon', 'Water_Carbon']
                      if col in year_data.columns]

        # Setup plotting
        plt.figure(figsize=(15, 5))

        # Plot 1: Carbon Pool Distribution (Bar Chart)
        plt.subplot(1, 2, 1)

        carbon_values = [year_data[col].values[0] for col in carbon_cols]
        plt.bar(carbon_cols, carbon_values, color=['green', 'brown', 'blue'])
        plt.title(f'Carbon Pool Distribution ({year})')
        plt.ylabel('Carbon (tonnes)')
        plt.grid(True, alpha=0.3)

        # Plot 2: Carbon Pool Distribution (Pie Chart)
        plt.subplot(1, 2, 2)

        plt.pie(carbon_values, labels=carbon_cols, autopct='%1.1f%%', colors=['green', 'brown', 'blue'])
        plt.title(f'Carbon Pool Proportion ({year})')

        plt.tight_layout()

        return plt

    def run_seasonal_analysis(self, year):
        """Run a seasonal analysis of carbon sequestration for a given year"""
        print(f"Running seasonal analysis for {year}...")

        # Create seasonal composites
        seasonal_composites = self.create_seasonal_composites(year)

        # Storage for results
        seasonal_results = []
        seasonal_maps = {}

        # Process each season
        for season, composite in seasonal_composites.items():
            if composite is None:
                print(f"No data available for {season} {year}, skipping.")
                continue

            print(f"Processing {season} {year}...")

            # Estimate carbon pools for this season
            carbon_pools, mangrove_mask = self.estimate_carbon_pools(composite, year)

            # Calculate area statistics
            area_stats = self.calculate_area_statistics(mangrove_mask, year)
            area_stats['season'] = season

            # Calculate carbon statistics
            carbon_stats = self.calculate_carbon_statistics(carbon_pools, year)

            # Combine statistics
            season_result = {**area_stats, **carbon_stats, 'season': season}
            seasonal_results.append(season_result)

            # Create visualization map
            map_widget = geemap.Map()
            map_widget.centerObject(self.sundarbans, 10)

            # Add base layers
            map_widget.addLayer(self.sundarbans, {'color': 'red'}, "Sundarbans Region")

            # Add carbon layer
            viz_params = {'min': 0, 'max': 500, 'palette': ['yellow', 'orange', 'darkgreen']}
            map_widget.addLayer(carbon_pools.select('Total_Carbon'), viz_params, f"Total Carbon ({season} {year})")

            # Add a legend
            map_widget.add_colorbar(viz_params, label="Total Carbon (tC/ha)")

            # Add info panel
            stats_html = f"""
            <div style="background-color: rgba(255, 255, 255, 0.8); padding: 10px; border-radius: 5px;">
                <h3>Carbon Analysis for {season.capitalize()} {year}</h3>
                <p><b>Mangrove Area:</b> {area_stats['mangrove_area_ha']:,.0f} hectares
                   ({area_stats['mangrove_percentage']:.1f}%)</p>
                <p><b>Total Carbon Stock:</b> {area_stats['total_carbon_stock']:,.0f} tonnes C</p>
                <p><b>Annual CO2 Absorption:</b> {area_stats['total_co2_seq']:,.0f} tonnes CO2/year</p>
            </div>
            """
            map_widget.add_html(stats_html, position="bottomright")

            seasonal_maps[season] = map_widget

        # Convert results to DataFrame
        seasonal_df = pd.DataFrame(seasonal_results)

        # Store results
        self.results['seasonal_analysis'] = {
            'year': year,
            'seasonal_df': seasonal_df,
            'seasonal_maps': seasonal_maps
        }

        # Create seasonal comparison chart
        if len(seasonal_df) > 0:
            plt.figure(figsize=(15, 10))

            # Plot 1: Mangrove Area by Season
            plt.subplot(2, 2, 1)
            sns.barplot(data=seasonal_df, x='season', y='mangrove_area_ha')
            plt.title(f'Mangrove Area by Season ({year})')
            plt.xlabel('Season')
            plt.ylabel('Area (hectares)')
            plt.xticks(rotation=45)
            plt.grid(True, alpha=0.3)

            # Plot 2: Total Carbon Stock by Season
            plt.subplot(2, 2, 2)
            sns.barplot(data=seasonal_df, x='season', y='total_carbon_stock')
            plt.title(f'Total Carbon Stock by Season ({year})')
            plt.xlabel('Season')
            plt.ylabel('Carbon Stock (tonnes C)')
            plt.xticks(rotation=45)
            plt.grid(True, alpha=0.3)

            # Plot 3: Biomass Carbon by Season
            plt.subplot(2, 2, 3)
            if 'Biomass_Carbon' in seasonal_df.columns:
                sns.barplot(data=seasonal_df, x='season', y='Biomass_Carbon')
                plt.title(f'Biomass Carbon by Season ({year})')
                plt.xlabel('Season')
                plt.ylabel('Biomass Carbon (tonnes C)')
                plt.xticks(rotation=45)
                plt.grid(True, alpha=0.3)

            # Plot 4: CO2 Absorption by Season
            plt.subplot(2, 2, 4)
            sns.barplot(data=seasonal_df, x='season', y='total_co2_seq')
            plt.title(f'CO2 Absorption by Season ({year})')
            plt.xlabel('Season')
            plt.ylabel('CO2 Absorption (tonnes/year)')
            plt.xticks(rotation=45)
            plt.grid(True, alpha=0.3)

            plt.tight_layout()

            self.results['seasonal_analysis']['chart'] = plt

        return self.results['seasonal_analysis']

    def create_dashboard(self):
        """Create a comprehensive dashboard with interactive widgets"""
        if 'historical_trends' not in self.results or 'carbon_pool_collection' not in self.results:
            print("Error: Historical analysis must be performed first.")
            return None

        # Get available years
        years = sorted(self.results['historical_trends']['year'].unique())

        # Create widgets
        year_dropdown = widgets.Dropdown(
            options=years,
            value=years[-1],
            description='Year:',
            style={'description_width': 'initial'}
        )

        carbon_type_dropdown = widgets.Dropdown(
            options=['Total_Carbon', 'Biomass_Carbon', 'Soil_Carbon', 'Water_Carbon'],
            value='Total_Carbon',
            description='Carbon Type:',
            style={'description_width': 'initial'}
        )

        # Create output area
        output = widgets.Output()

        # Define update function
        def update_map(change):
            with output:
                output.clear_output()
                year = year_dropdown.value
                carbon_type = carbon_type_dropdown.value

                map_widget = self.visualize_carbon_map(year=year, carbon_type=carbon_type)
                display(map_widget)

        # Register callbacks
        year_dropdown.observe(update_map, names='value')
        carbon_type_dropdown.observe(update_map, names='value')

        # Create dashboard layout
        dashboard = widgets.VBox([
            widgets.HBox([year_dropdown, carbon_type_dropdown]),
            output
        ])

        # Initial display
        update_map(None)

        return dashboard

    def create_summary_report(self):
        """Create a comprehensive summary report of findings"""
        if 'historical_trends' not in self.results:
            print("Error: Analysis must be performed first.")
            return None

        historical_data = self.results['historical_trends']

        # Basic statistics
        first_year = historical_data['year'].min()
        last_year = historical_data['year'].max()

        first_year_data = historical_data[historical_data['year'] == first_year].iloc[0]
        last_year_data = historical_data[historical_data['year'] == last_year].iloc[0]

        # Calculate changes
        mangrove_change = last_year_data['mangrove_area_ha'] - first_year_data['mangrove_area_ha']
        mangrove_change_percent = (mangrove_change / first_year_data['mangrove_area_ha']) * 100

        carbon_change = last_year_data['total_carbon_stock'] - first_year_data['total_carbon_stock']
        carbon_change_percent = (carbon_change / first_year_data['total_carbon_stock']) * 100

        report = {
            'period': f"{first_year} to {last_year}",
            'initial_mangrove_area': first_year_data['mangrove_area_ha'],
            'final_mangrove_area': last_year_data['mangrove_area_ha'],
            'mangrove_change': mangrove_change,
            'mangrove_change_percent': mangrove_change_percent,
            'initial_carbon_stock': first_year_data['total_carbon_stock'],
            'final_carbon_stock': last_year_data['total_carbon_stock'],
            'carbon_change': carbon_change,
            'carbon_change_percent': carbon_change_percent,
            'co2_absorption_rate': last_year_data['total_co2_seq']
        }

        # Create future projections summary if available
        if 'future_predictions' in self.results:
            future_data = self.results['future_predictions']
            last_prediction_year = future_data['year'].max()
            last_prediction = future_data[future_data['year'] == last_prediction_year].iloc[0]

            future_mangrove_change = last_prediction['mangrove_area_ha'] - last_year_data['mangrove_area_ha']
            future_mangrove_change_percent = (future_mangrove_change / last_year_data['mangrove_area_ha']) * 100

            future_carbon_change = last_prediction['total_carbon_stock'] - last_year_data['total_carbon_stock']
            future_carbon_change_percent = (future_carbon_change / last_year_data['total_carbon_stock']) * 100

            report.update({
                'projection_end_year': last_prediction_year,
                'projected_mangrove_area': last_prediction['mangrove_area_ha'],
                'projected_mangrove_change': future_mangrove_change,
                'projected_mangrove_change_percent': future_mangrove_change_percent,
                'projected_carbon_stock': last_prediction['total_carbon_stock'],
                'projected_carbon_change': future_carbon_change,
                'projected_carbon_change_percent': future_carbon_change_percent,
                'projected_co2_absorption_rate': last_prediction['total_co2_seq']
            })

        # Store report
        self.results['summary_report'] = report

        # Print text summary
        print("\n================ SUNDARBANS CARBON ANALYSIS SUMMARY ================\n")
        print(f"Analysis Period: {report['period']}")
        print("\n--- MANGROVE EXTENT ---")
        print(f"Initial Mangrove Area ({first_year}): {report['initial_mangrove_area']:,.2f} hectares")
        print(f"Final Mangrove Area ({last_year}): {report['final_mangrove_area']:,.2f} hectares")
        print(f"Change: {report['mangrove_change']:,.2f} hectares ({report['mangrove_change_percent']:.2f}%)")

        print("\n--- CARBON STOCKS ---")
        print(f"Initial Carbon Stock ({first_year}): {report['initial_carbon_stock']:,.2f} tonnes C")
        print(f"Final Carbon Stock ({last_year}): {report['final_carbon_stock']:,.2f} tonnes C")
        print(f"Change: {report['carbon_change']:,.2f} tonnes C ({report['carbon_change_percent']:.2f}%)")
        print(f"Current CO2 Absorption Rate: {report['co2_absorption_rate']:,.2f} tonnes CO2/year")

        if 'future_predictions' in self.results:
            print(f"\n--- FUTURE PROJECTIONS (by {report['projection_end_year']}) ---")
            print(f"Projected Mangrove Area: {report['projected_mangrove_area']:,.2f} hectares")
            print(f"Projected Change from {last_year}: {report['projected_mangrove_change']:,.2f} hectares ({report['projected_mangrove_change_percent']:.2f}%)")
            print(f"Projected Carbon Stock: {report['projected_carbon_stock']:,.2f} tonnes C")
            print(f"Projected Change from {last_year}: {report['projected_carbon_change']:,.2f} tonnes C ({report['projected_carbon_change_percent']:.2f}%)")
            print(f"Projected CO2 Absorption Rate: {report['projected_co2_absorption_rate']:,.2f} tonnes CO2/year")

        print("\n==================================================================\n")

        return report

    def run_full_analysis(self, start_year=2018, end_year=2024, prediction_years=5):
        """Run a complete analysis pipeline from start to finish"""
        print("Starting full Sundarbans carbon sink analysis...")

        # 1. Analyze historical trends
        self.analyze_historical_trends(start_year, end_year)

        # 2. Predict future carbon
        self.predict_future_carbon(prediction_years)

        # 3. Detect changes between first and last historical year
        self.detect_changes(start_year, end_year)

        # 4. Run seasonal analysis for the most recent year
        self.run_seasonal_analysis(end_year)

        # 5. Create comprehensive results summary
        self.create_summary_report()

        print("Analysis complete!")

        return self.results

    def display_results(self):
        """Display key results and visualizations"""
        # Check if analysis has been run
        if not self.results:
            print("No results to display. Please run the analysis first.")
            return

        # 1. Display summary report
        if 'summary_report' in self.results:
            print("\n----- SUMMARY REPORT -----")
            self.create_summary_report()

        # 2. Display carbon map for most recent year
        if 'historical_trends' in self.results:
            print("\n----- CARBON DISTRIBUTION MAP -----")
            last_year = self.results['historical_trends']['year'].max()
            carbon_map = self.visualize_carbon_map(year=last_year)
            display(carbon_map)

        # 3. Display change map
        if 'change_detection' in self.results:
            print("\n----- CARBON CHANGE MAP -----")
            change_map = self.visualize_change_map()
            display(change_map)

        # 4. Display time series charts
        if 'complete_timeline' in self.results:
            print("\n----- TIME SERIES ANALYSIS -----")
            time_series_plt = self.create_time_series_charts()
            display(time_series_plt.figure)

        # 5. Display comparative charts for most recent year
        if 'historical_trends' in self.results:
            print("\n----- CARBON POOL DISTRIBUTION -----")
            last_year = self.results['historical_trends']['year'].max()
            comparative_plt = self.create_comparative_charts(year=last_year)
            display(comparative_plt.figure)

        # 6. Display seasonal analysis results if available
        if 'seasonal_analysis' in self.results:
            print("\n----- SEASONAL ANALYSIS -----")
            seasonal = self.results['seasonal_analysis']

            # Display seasonal chart
            if 'chart' in seasonal:
                display(seasonal['chart'].figure)

            # Display seasonal data
            if 'seasonal_df' in seasonal:
                print("\nSeasonal Data:")
                display(seasonal['seasonal_df'])

            # Only display the first seasonal map to avoid cluttering
            if 'seasonal_maps' in seasonal and seasonal['seasonal_maps']:
                print("\nSeasonal Map (first season):")
                first_season = list(seasonal['seasonal_maps'].keys())[0]
                display(seasonal['seasonal_maps'][first_season])


# Example usage
def run_sundarbans_analysis(start_year=2018, end_year=2024, prediction_years=5):
    """Run the complete Sundarbans carbon sink analysis and visualize results"""
    # Initialize the analyzer
    analyzer = SundarbansAnalysis()

    # Run full analysis
    analyzer.run_full_analysis(start_year, end_year, prediction_years)

    # Display results
    analyzer.display_results()

    # Create interactive dashboard
    print("\n----- INTERACTIVE DASHBOARD -----")
    print("Use the dropdown menus to explore carbon distribution for different years and carbon pools.")
    dashboard = analyzer.create_dashboard()
    display(dashboard)

    return analyzer



# Main execution block
if __name__ == "__main__":
    print("============================================================")
    print("   SUNDARBANS MANGROVE CARBON SINK ANALYSIS SYSTEM")
    print("============================================================")
    print("This system analyzes carbon sequestration in the Sundarbans")
    print("mangrove forest using Google Earth Engine and Python.\n")

    # Run the analysis
    analyzer = run_sundarbans_analysis(start_year=2018, end_year=2024, prediction_years=5)

    print("\nAnalysis complete!")
    print("============================================================")
    print("Use the interactive dashboard above to explore the results.")
    print("============================================================")
```
